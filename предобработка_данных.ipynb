```python
import pandas as pd
import re
from datetime import datetime

# Функция для извлечения даты из URL
def extract_date_from_url(url):
    try:
        # Ищем паттерн /год/месяц/день/ в URL
        match = re.search(r'/(\d{4})/([a-z]{3})/(\d{1,2})/', url.lower())
        if match:
            year, month, day = match.groups()
            month_num = datetime.strptime(month, "%b").month
            return pd.to_datetime(f"{year}-{month_num}-{day}")
    except:
        pass
    return pd.NaT

grd['date'] = grd['url'].apply(extract_date_from_url)

start_date = pd.to_datetime('2023-11-03')
end_date = pd.to_datetime('2024-11-07')

grd = grd[(grd['date'] >= start_date) & (grd['date'] <= end_date)].copy()

grd["media"] = "grd"

grd
```

```python
import pandas as pd
from datetime import datetime

wpp['date'] = pd.to_datetime(wpp['date']).dt.date

wpp['date'] = pd.to_datetime(wpp['date'])

start_date = pd.to_datetime('2023-11-03')
end_date = pd.to_datetime('2024-11-07')

wpp = wpp[(wpp['date'] >= start_date) & (wpp['date'] <= end_date)]

wpp["media"] = "wpp"

wpp
```

```python
def safe_concat(dfs):
    all_columns = set().union(*[set(df.columns) for df in dfs])
    
    for df in dfs:
        for col in all_columns - set(df.columns):
            df[col] = pd.NA
    
    return pd.concat(dfs, ignore_index=True)

dataframes = [grd, wpp, nyp, fxn]

data = safe_concat(dataframes)

data['ideology'] = data['media'].apply(lambda x: 'L' if x in ['grd', 'wpp'] else 'R')

data
```

```python
def count_words(text):
    if pd.isna(text):
        return 0
    return len(str(text).split())

data['word_count'] = data['text'].apply(count_words)

data = data[data['word_count'] >= 50].copy()

data.drop('word_count', axis=1, inplace=True)

data
```

```python
from collections import defaultdict

name_counts = defaultdict(int)

names = [str(name) for name in names if name is not None]

w_names = pd.DataFrame()

for name in names:
    filtered = data[data["text"].str.contains(name, na=False, regex=False)]
    
    count = len(filtered)
    name_counts[name] = count
    
    w_names = pd.concat([w_names, filtered], ignore_index=True)

w_names = w_names.drop_duplicates()

print("Количество упоминаний каждого имени (по убыванию):")
for name, count in sorted(name_counts.items(), key=lambda item: item[1], reverse=True):
    print(f"{name}: {count} упоминаний")
```

```python
import pandas as pd
import re
from collections import defaultdict

expanded_data = pd.DataFrame()

for idx, row in w_names.iterrows():
    text = str(row['text']) 
    
    mention_counter = 1 
    
    for name in names:
        matches = re.finditer(re.escape(str(name)), text, re.IGNORECASE)
        
        for match in matches:
            words = text.split()
            
            char_pos = match.start()
            pre_text = text[:char_pos]
            pre_words = pre_text.split()
            target_word_pos = len(pre_words)
            
            start = max(0, target_word_pos - 35)
            end = min(len(words), target_word_pos + 35 + 1)
            context_text = ' '.join(words[start:end])
            
            new_row = row.copy()
            new_row['original_title'] = new_row['title']
            new_row['title'] = f"{new_row['title']}_{mention_counter}"
            new_row['text_fragment'] = context_text
            new_row['matched_name'] = name
            
            expanded_data = pd.concat([expanded_data, pd.DataFrame([new_row])], ignore_index=True)
            
            mention_counter += 1

data_for_model = expanded_data.drop_duplicates(subset=['text_fragment'])

original_columns = w_names.columns.tolist()
new_columns = ['original_title', 'matched_name', 'text_fragment']
all_columns = original_columns + [col for col in new_columns if col not in original_columns]

data_for_model = expanded_data[all_columns]

print(f"Всего найдено {len(data_for_model)} уникальных контекстных фрагментов")
print(f"Исходные колонки сохранены: {w_names.columns.tolist()}")
print("\nПример данных:")
print(data_for_model.head(3)[['url', 'date', 'title', 'matched_name', 'text_fragment']])
```
